```python
import girder_client
import os

def maybe_mkdir(path):
    a, b = os.path.split(path)
    if a != "":
        maybe_mkdir(a)
    if not os.path.exists(path):
        os.mkdir(path)

maybe_mkdir("SpineSegmentationData/npy")
    
maybe_mkdir("SpineSegmentationData/training/ultrasounds")
maybe_mkdir("SpineSegmentationData/training/segmentations")
maybe_mkdir("SpineSegmentationData/testing/ultrasounds")
maybe_mkdir("SpineSegmentationData/testing/segmentations")
    
    
gc = girder_client.GirderClient(apiUrl='https://pocus.cs.queensu.ca/api/v1')

if not os.path.exists("SpineSegmentationData/npy/ultrasound-test.npy"):
    print("Downloading data. This may take several minutes")
    gc.downloadFolderRecursive("5da9e563d9e6a3be02d012b1", "SpineSegmentationData/npy/")
```

```python
import numpy as np

import skimage.io
for fname in sorted(os.listdir("SpineSegmentationData/npy/")):
    print("==============")
    print(fname)
    if "test" in fname:
        prefix="testing"
    else:
        prefix="training"
        
    if "ultrasound" in fname:
        type_ = "ultrasounds"
    else:
        type_ = "segmentations"
    base = os.path.join("SpineSegmentationData", prefix, type_, fname[:-4])
    print(base)
    array = np.load("SpineSegmentationData/npy/" + fname)
    array = array * 255
    array = array.astype(np.uint8)
    print(np.max(array), np.min(array))
    for i, image in enumerate(array):
        skimage.io.imsave(base + str(i) + ".png", image[:, :, 0], check_contrast=False)
```
